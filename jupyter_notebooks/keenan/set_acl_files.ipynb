```python
def make_patch(i, accession, status):
    complete = False
    print('{} {} {}'.format(i, accession, status), end=' ')
    res = requests.patch(
        'https://www.encodeproject.org/files/{}/@@set_status?block_children=true&update=true'.format(accession),
        auth=('E4D25Y5U', 'ahnov7i4womkfgut'), json={'status': '{}'.format(status)}
    )
    feedback = (accession, res)
    if res.status_code != 200:
        print('error')
        patch_errors.append(feedback)
    else:
        print('success')
        patch_successes.append(feedback)
        complete = True
    return complete
        

def get_status(accession):
    complete = False
    print('Getting status for {}'.format(accession), end=' ')
    res = requests.get(
        'https://www.encodeproject.org/files/{}/'.format(accession),
        auth=('E4D25Y5U', 'ahnov7i4womkfgut')
    )
    status = res.json().get('status')
    feedback = (accession, res)
    if res.status_code != 200:
        print('error')
        get_errors.append(feedback)
    else:
        print(status)
        get_successes.append(feedback)
        complete = True
    return status, complete
        
        
def run_update(accessions):
    for i, accession in enumerate(accessions):
        status, get_status_complete = get_status(accession)
        if not get_status_complete:
            continue
        if status not in ['released', 'archived', 'revoked']:
            continue
        patch_complete = make_patch(i, accession, status)
```
